```r
bmAtlas <- readRDS("RDS/GSE253355_Normal_Bone_Marrow_Atlas_Seurat_SB_v2.rds")
bmAtlas <- RunUMAP(bmAtlas, dims = 1:30, seed.use = 42, return.model = TRUE)

sn_obj <- bm_cm
## Re-normalize with LogNorm
DefaultAssay(sn_obj) <- "Nanostring"
sn_obj <- NormalizeData(sn_obj) %>% FindVariableFeatures(nfeatures = 300) %>%
    ScaleData() %>% RunPCA() 
## Label transfer
sn.anchors <- FindTransferAnchors(reference = bmAtlas, query = sn_obj, dims = 1:30, reference.reduction = "pca")
sn_obj <- MapQuery(anchorset = sn.anchors, reference = bmAtlas, query = sn_obj,
    refdata = bmAtlas$cluster_anno_l2, reference.reduction = "pca", reduction.model = "umap")
sn_obj$pred_cell_type <- ifelse(sn_obj$predicted.id.score > 0.6, sn_obj$predicted.id, "Not-mapped")
saveRDS(sn_obj, "/diskmnt/Projects/SenNet_analysis/Main.analysis/bm/CosMx/MSH_samples/MB22MSH00608A1/MSH608_labelTransfer_to_TanRefUMAP.rds")
## Plot QC
library(scales)
my_color_palette <- hue_pal()(length(unique(sn_obj$pred_cell_type)))
my_color_palette[which(sort(unique(sn_obj$pred_cell_type))=="Not-mapped")] <- '#DBDED9'
pdf("/diskmnt/Projects/SenNet_analysis/Main.analysis/bm/CosMx/MSH_samples/MB22MSH00608A1/MSH608_labelTransfer_UMAP_QC.pdf", width = 16, height = 8)
p1 <- DimPlot(bmAtlas, reduction = "umap", group.by = "cluster_anno_l2", label = TRUE, label.size = 3,
    repel = TRUE) + NoLegend() + ggtitle("snRNA annotations") + coord_fixed(ratio=1)
p2 <- DimPlot(sn_obj, reduction = "ref.umap", group.by = "pred_cell_type", label = TRUE, cols = my_color_palette,
    label.size = 3, repel = TRUE) + ggtitle("Label transfer prediction") + coord_fixed(ratio=1)
print(p1+p2)
print(DimPlot(sn_obj, group.by = "pred_cell_type", cols = my_color_palette, reduction = "umap.30PC", label = TRUE)+ coord_fixed(ratio=1))
print(DimPlot(sn_obj, group.by = "seurat_clusters", reduction = "umap.30PC", label = TRUE)+ coord_fixed(ratio=1))
dev.off()
```
